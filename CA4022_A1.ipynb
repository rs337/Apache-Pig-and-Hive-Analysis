```python
import pandas as pd
column_names = ['Name','Calories', 'Fat', 'Carb', 'Fiber', 'Protein']
df = pd.read_csv('starbucks_food', names=column_names, index_col=False)
```

```python
import matplotlib.pyplot as plt
top_proteins = df.sort_values(by='Protein', ascending=False).head(5)
plt.figure(figsize=(10, 6))
plt.bar(top_proteins['Name'], top_proteins['Protein'])
plt.xlabel('Food Name')
plt.ylabel('Protein (g)')
plt.title('Top 5 Foods with the Highest Protein Content')
plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels for better visibility
plt.tight_layout()
plt.show()
df.head()
```

```python
column_names = ['BevCat', 'Bev', 'Bev_Prep', 'Calories', 'Fat', 'Sodium', 'Carb', 'Fibre', 'Protein']
df = pd.read_csv('starbucks_menu', names=column_names, index_col=False, skiprows=1)
df.head(5)
```

```python
print(df.columns)
```

```python
category_stats = df.groupby('BevCat')[['Calories', 'Protein']].mean()
top_3_categories = category_stats.sort_values(by='Protein', ascending=False).head(3)

plt.figure(figsize=(10, 6))
for category in top_3_categories.index:
    data = top_3_categories.loc[category]
    plt.scatter(data['Calories'], data['Protein'], label=category)

plt.xlabel('Average Calories')
plt.ylabel('Average Protein')
plt.title('Average Calories vs. Average Protein for Top 3 Highest Protein Food Categories')
plt.legend()
plt.show()
```
